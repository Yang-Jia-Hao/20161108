```python
class Option:
    def __init__(self, K, Interval, Premium_Call, Premium_Put):
        import numpy as np
        self.K = K
        self.Interval = Interval #Range
        self.Premium_Call = Premium_Call
        self.Premium_Put = Premium_Put
        
    
    def Call(self):
        import numpy as np
        ST = np.arange(self.K - self.Interval, self.K + self.Interval) #S0~ST期 X軸
        #payoff 計算 向量化運算
        Payoff_LongCall = np.maximum(ST - self.K, 0) - self.Premium_Call #Long Call 買買權
        Payoff_ShortCall = - Payoff_LongCall                   #Short Call 賣買權
        import matplotlib.pyplot as plt #簡化叫plt
        plt.figure
        plt.plot(ST, Payoff_LongCall)
        plt.plot(ST, Payoff_ShortCall)
        plt.plot(ST, Payoff_LongCall, label=u'$+C$' ,color="red" ,linewidth=2) # 加上label參數添加圖例
        plt.plot(ST, Payoff_ShortCall, label=u'$-C$',color="blue" ,linewidth=8 )
        plt.xlabel("123")
        plt.ylabel('456')
        plt.grid()   #網格
        plt.legend() #加這個我的圖例才出來
        plt.axvline(K, color='black' ,ls='dashed')
        plt.axvline(K + Premium_Put, color='black', ls='dashed')
        plt.axhline(0, color='black', ls='dashed')   #0虛線
        
    def Put(self):
        %matplotlib notebook
        ST = np.arange(self.K - self.Interval, self.K + self.Interval) #S0~ST期 X軸
        Payoff_LongPut = np.maximum(K - ST, 0) - Premium_Put
        Payoff_ShortPut = - Payoff_LongPut
        import matplotlib.pyplot as plt #簡化叫plt
        plt.figure
        plt.plot(ST, Payoff_LongPut)
        plt.plot(ST, Payoff_ShortPut)
        plt.plot(ST, Payoff_LongPut, label=u'$+P$' ,color="red",linewidth=2) # 加上label參數添加圖例
        plt.plot(ST, Payoff_ShortPut, label=u'$-P$',color="blue",linewidth=8)
        plt.axvline(K, color='black' ,ls='dashed')
        plt.axvline(K-Premium_Put, color='black', ls='dashed')
        plt.axhline(0, color='black', ls='dashed')   #0虛線
        plt.legend()
```

```python
K = 9100
Interval = 500 #Range
Premium_Call = 179
Premium_Put = 185

C1 = Option(K, Interval, Premium_Call, Premium_Put)
C1.Call()
```
